```python
from datetime import datetime
import pandas as pd
import numpy as np
import mplfinance as mpf
import talib
import matplotlib.pyplot as plt

RawData = pd.read_csv('2327 .csv', index_col = 'Date')

RawData.index = pd.DatetimeIndex(RawData.index)
RawData = RawData.loc['2019/01/02':'2020/10/30',:]

K,D = talib.STOCH(RawData['High'].values, RawData['Low'].values, RawData['Close'].values)

AddPlotPanel_1 = [ mpf.make_addplot(K, color='g', panel=1),
                   mpf.make_addplot(D, color='b', panel=1),
                 ]

Result = mpf.plot(RawData, type = 'candle', mav = [10, 30], style = 'charles',volume = True, figscale = 2, 
         main_panel = 0, volume_panel = 2, addplot = AddPlotPanel_1, returnfig = True)
```
